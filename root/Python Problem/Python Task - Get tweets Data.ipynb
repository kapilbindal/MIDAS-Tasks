### Fetching tweets of a user from Twitter API and displaying the relevant data as per the guidelines

### Import Libraries

```python
import tweepy
import json
import pandas as pd
import numpy as np
```

### Defining Keys  and Authorising 

```python
consumerKey = "XXXXXXXXXXXXX"
consumerSecret = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"
accessToken = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"
accessSecret = "XXXXXXXXXXXXXXXXXXXXXXXXXXX"
```

```python
auth = tweepy.OAuthHandler(consumerKey,consumerSecret)
auth.set_access_token(accessToken,accessSecret)
api = tweepy.API(auth)
```

### Method to Fetch Tweets

```python
def getTweets(username):
    tweets = api.user_timeline(id =username, tweet_mode = 'extended',count = 150) # using tweepy.Cursor() would give at max 329 tweets
    return tweets
```

### Saving data to JSON File

```python
tweets = getTweets("midasIIITD")
tweetsDump = []
for tweet in tweets:
    tweetsDump.append(tweet._json)
```

```python
with open('tweets.json', 'w') as outfile:
    json.dump(tweetsDump, outfile)
```

### Reading data from JSON file

```python
with open('tweets.json') as f:
    tweetsData = json.load(f)
```

```python
print(tweetsData[0].keys()) # All the keys in each tweet
```

### Making Lists to store the Counts

```python
texts = []
dateTime = []
likesCount = []
retweetsCount = []
photosCount = []
```

```python
for tweet in tweetsData:
    texts.append(tweet['full_text'])
    dateTime.append(tweet['created_at'])
    likesCount.append(tweet['favorite_count'])
    retweetsCount.append(tweet['retweet_count'])
    numImages = 0;
    if 'media' in tweet['entities']:
        numImages = int(len(tweet['extended_entities']['media'])) # this gives all the attached media with the tweet
    if(numImages > 0):
        photosCount.append(numImages)
    else:
        photosCount.append('None')   
```

### Tabular Data

```python
df = pd.DataFrame()
```

```python
df['Text'] = texts
df['Date and Time'] = dateTime
df['Likes'] = likesCount
df['Retweets'] = retweetsCount
df['No. of Images'] = photosCount
```
